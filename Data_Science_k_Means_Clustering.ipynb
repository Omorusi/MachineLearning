```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/content/Online Purchase.csv")
data.head()
```

```python
# Get the number of rows and columns
rows, columns = data.shape

print(f"Dataset has {rows} rows and {columns} columns.")
```

```python
data_cleaned = data.dropna()

# Get the number of rows and columns
rows, columns = data_cleaned.shape

print(f"Dataset has {rows} rows and {columns} columns.")

```

```python
plt.scatter(data_cleaned['Age'], data_cleaned['Salary'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of Age vs Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter dataset to remove high salaries & unrealistic ages
filtered_data = data_cleaned[(data_cleaned['Salary'] <= 190000) & (data_cleaned['Age'] <= 100)]

# Scatter plot
plt.scatter(filtered_data['Age'], filtered_data['Salary'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of Age vs Salary (Filtered)')

plt.show()
```

```python
km =  KMeans(n_clusters=4)
km
```

```python
y_predicted =  km.fit_predict(data_cleaned[['Age','Salary']])
y_predicted
```

```python
data_cleaned['cluster'] = y_predicted
data_cleaned.head()
```

```python
y_predicted =  km.fit_predict(data_cleaned[['Age','Salary']])
y_predicted
```

```python
data1 = data_cleaned[data_cleaned['cluster']==0]
data2 = data_cleaned[data_cleaned['cluster']==1]
data3 = data_cleaned[data_cleaned['cluster']==2]
plt.scatter(data1['Age'],data1['Salary'],color='green')
plt.scatter(data2['Age'],data2['Salary'],color='red')
plt.scatter(data3['Age'],data3['Salary'],color='blue')


plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
